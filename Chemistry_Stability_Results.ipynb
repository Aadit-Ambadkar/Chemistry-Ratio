```
%matplotlib inline
from fastai.basics import *
from fastai.tabular import*
```

```
path=Path("/content/data")
path
```

```
data=pd.read_csv(path/"Discovering the Stability Formula.csv", header=None, index_col=False)
data.head()
```

```
x=[]
```

```
y=[]
```

```
for i in range(len(data)):
  x_val = data[0][i]
  y_val = data[1][i]
  x.append([float(x_val**2), float(x_val), float(1)])
  y.append(y_val-x_val)
```

```
x[-5:], y[-5:]
```

```
x=tensor(x)
y=tensor(y)
```

```
manual_bound = tensor([0.0083, 0.8333, 0])
lower_bound = tensor([0.0066, 0.9470, 0.8417])
upper_bound = tensor([0.0052, 1.1165, 0.8210])
final_bound = tensor([0.0059, 1.0321, 0.9998])
```

```
plt.figure(figsize=(20,32))
plt.subplot(2, 2, 1)
plt.scatter(x[:,1],y)
plt.scatter(x[:,1],x@upper_bound.detach())
plt.subplot(2, 2, 2)
plt.scatter(x[:,1],y)
plt.scatter(x[:,1],x@final_bound.detach())
plt.subplot(2, 2, 3)
plt.scatter(x[:,1],y)
plt.scatter(x[:,1],x@lower_bound.detach())
plt.subplot(2, 2, 4)
plt.scatter(x[:,1],y)
plt.scatter(x[:,1],x@tensor(0.,1,1).detach())
```

```
plt.figure(figsize=(6,8))
plt.scatter(x[:,1],y,s=10)
```

```
plt.figure(figsize=(6,8))
plt.scatter(x[:,1],y,s=10)
plt.scatter(x[:,1],x@manual_bound.detach(),s=10)
```
